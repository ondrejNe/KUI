```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Classifiers 
cls = ['C1.dsv', 'C2.dsv', 'C3.dsv', 'C4.dsv', 'C5.dsv']
for cl in cls:
    # Load data
    classifier_outputs = pd.read_csv(f'classif_result_tables/{cl}', delimiter=',', header=None)
    ground_truth = pd.read_csv('classif_result_tables/GT.dsv', delimiter=',', header=None).to_numpy().flatten()

    # Prepare to store metrics
    tpr_list = []  # True positive rate list
    fpr_list = []  # False positive rate list
    tnr_list = []  # True negative rate list
    fnr_list = []  # False negative rate list
    alphas = []  # Store alpha indices for labeling


    # Calculate metrics for each alpha
    for i in range(classifier_outputs.shape[1]):
        predictions = classifier_outputs[i]
        TP = np.sum((predictions == 1) & (ground_truth == 1))
        FP = np.sum((predictions == 1) & (ground_truth == 0))
        TN = np.sum((predictions == 0) & (ground_truth == 0))
        FN = np.sum((predictions == 0) & (ground_truth == 1))
        
        TPR = TP / (TP + FN) if TP + FN != 0 else 0
        FPR = FP / (FP + TN) if FP + TN != 0 else 0

        TNR = TN / (TN + FP) if TN + FP != 0 else 0
        FNR = FN / (FN + TP) if FN + TP != 0 else 0
        
        tpr_list.append(TPR)
        fpr_list.append(FPR)
        tnr_list.append(TNR)
        fnr_list.append(FNR)
        alphas.append(f"{i}")

    # Plot ROC points
    plt.figure(figsize=(10, 8))
    plt.plot(fpr_list, tpr_list, color='blue')
    plt.scatter(fpr_list, tpr_list, color='red')

    # Label each point with its corresponding alpha
    for i, label in enumerate(alphas):
        plt.annotate(label, (fpr_list[i], tpr_list[i]), textcoords="offset points", xytext=(10,-10), ha='center')

    # Set custom ticks to match the FPR and TPR values
    plt.xticks(fpr_list, [f"{x:.2f}" for x in fpr_list])
    plt.yticks(tpr_list, [f"{y:.2f}" for y in tpr_list])

    # Limits and labels
    xlim = [min(fpr_list) - 0.05, max(fpr_list) + 0.05]
    ylim = [min(tpr_list) - 0.05, max(tpr_list) + 0.05]
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title(f'{cl} - Receiver Operating Characteristic (ROC) Curve')
    plt.show()

    # Generating labels from 0 to 49
    labels = list(range(50))  # Adjust the range if the list lengths are different

    # Printing labels
    print("Labels    : ", end='')
    print(*[f'{x:4}' for x in labels], sep=', ')

    # Printing TPR values with two decimal places
    print("TPR values: ", end='')
    print(*[f'{x:.2f}' for x in tpr_list], sep=', ')

    # Printing FPR values with two decimal places
    print("FPR values: ", end='')
    print(*[f'{x:.2f}' for x in fpr_list], sep=', ')
```
